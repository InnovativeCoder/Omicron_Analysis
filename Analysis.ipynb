```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import seaborn as sns
```

```python
#import data
data = pd.read_csv('./data/Nov 2021/11-01-2021.csv')
#Now we have collection CSVs
data
```

```python
#Constructing new data for total cases
new_data = data.iloc[:,7:9]
new_data["Deaths"]
#Total deaths on particular date all over the globe
# for i in new_data["Deaths"]
total_death = 0
total_confirmed_cases = 0
for i in range(0,len(new_data)):
    total_death+=new_data["Deaths"][i]
    total_confirmed_cases+=new_data["Confirmed"][i]
total_death, total_confirmed_cases
```

```python
#Constructing new data for total cases according to dates for Nov 2021
new_total_nov_data = {'Date':[], 'Confirmed Cases':[], 'Deaths':[]}
date_arr = []
confirmed_arr = []
deaths_arr = []
for i in range(1,10):
    data = pd.read_csv('./data/Nov 2021/11-0'+str(i)+'-2021.csv')
    print(str(i)+"/11/2021")
    date_arr.append(str(i)+"/11/2021")
    #Constructing new data for total cases
    new_data = data.iloc[:,7:9]
    new_data["Deaths"]
    #Total deaths on particular date all over the globe
    # for i in new_data["Deaths"]
    total_death = 0
    total_confirmed_cases = 0
    for i in range(0,len(new_data)):
        total_death+=new_data["Deaths"][i]
        total_confirmed_cases+=new_data["Confirmed"][i]
    print(total_death, total_confirmed_cases)
    confirmed_arr.append(total_confirmed_cases)
    deaths_arr.append(total_death)
    
for i in range(10,31):
    data = pd.read_csv('./data/Nov 2021/11-'+str(i)+'-2021.csv')
    print(str(i)+"/11/2021")
    date_arr.append(str(i)+"/11/2021")
    #Constructing new data for total cases
    new_data = data.iloc[:,7:9]
    new_data["Deaths"]
    #Total deaths on particular date all over the globe
    # for i in new_data["Deaths"]
    total_death = 0
    total_confirmed_cases = 0
    for i in range(0,len(new_data)):
        total_death+=new_data["Deaths"][i]
        total_confirmed_cases+=new_data["Confirmed"][i]
    print(total_death, total_confirmed_cases)
    confirmed_arr.append(total_confirmed_cases)
    deaths_arr.append(total_death)
    
new_total_nov_data["Date"] = date_arr
new_total_nov_data["Confirmed Cases"] = confirmed_arr
new_total_nov_data["Deaths"] = deaths_arr
new_total_nov_dataframe = pd.DataFrame(new_total_nov_data)
```

```python
#Constructing new data for total cases according to dates for Jan 2022
new_total_jan_data = {'Date':[], 'Confirmed Cases':[], 'Deaths':[]}
date_arr = []
confirmed_arr = []
deaths_arr = []
for i in range(1,7):
    data = pd.read_csv('./data/Jan 2022/01-0'+str(i)+'-2022.csv')
    print(str(i)+"/01/2022")
    date_arr.append(str(i)+"/01/2022")
    #Constructing new data for total cases
    new_data = data.iloc[:,7:9]
    new_data["Deaths"]
    #Total deaths on particular date all over the globe
    # for i in new_data["Deaths"]
    total_death = 0
    total_confirmed_cases = 0
    for i in range(0,len(new_data)):
        total_death+=new_data["Deaths"][i]
        total_confirmed_cases+=new_data["Confirmed"][i]
    print(total_death, total_confirmed_cases)
    confirmed_arr.append(total_confirmed_cases)
    deaths_arr.append(total_death)
    

new_total_jan_data["Date"] = date_arr
new_total_jan_data["Confirmed Cases"] = confirmed_arr
new_total_jan_data["Deaths"] = deaths_arr
new_total_jan_dataframe = pd.DataFrame(new_total_jan_data)
```

```python
# Total Data

total_cases_dataframe = new_total_nov_dataframe

total_cases_dataframe = total_cases_dataframe.append(new_total_dec_dataframe, ignore_index=True)
total_cases_dataframe = total_cases_dataframe.append(new_total_jan_dataframe, ignore_index=True)
total_cases_dataframe
```

```python
plt.plot(total_cases_dataframe["Date"], total_cases_dataframe["Deaths"], color='red')
#legend
one = mpatches.Patch(facecolor='#ff0000', label='No. of Deaths', linewidth = 0.5, edgecolor = 'black')
legend = plt.legend(handles=[one], loc = 4, fontsize = 'small', fancybox = True)
frame = legend.get_frame() #sets up for color, edge, and transparency
frame.set_facecolor('#b4aeae') #color of legend
frame.set_edgecolor('black') #edge color of legend
frame.set_alpha(1) #deals with transparency

plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
plt.title('Total no of deaths daily')
plt.xlabel('Dates From Nov 2021 to Jan 2022')
plt.ylabel('Deaths')
plt.savefig("Plot-Total-no-of-deaths")
plt.show()
```
